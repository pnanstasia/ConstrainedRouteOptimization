```python
import pandas as pd
from pathlib import Path
import osmnx as ox
import matplotlib.pyplot as plt
```

```python
def create_colored_map_osmnx(csv_path):
    df = pd.read_csv(csv_path)
    df["lan"] = pd.to_numeric(df["lan"], errors="coerce")
    df["lot"] = pd.to_numeric(df["lot"], errors="coerce")
    df = df.dropna(subset=["lan", "lot"]).copy()

    color_map = {
        "ATM": "#1F39FF",
        "TC": "#FF6D1F",
        "TOBO": "#543BE3",
        "TT": "#E73671",
    }

    pad = 0.01
    north = df["lan"].max() + pad
    south = df["lan"].min() - pad
    east = df["lot"].max() + pad
    west = df["lot"].min() - pad

    graph = ox.graph_from_bbox(
        north=north,
        south=south,
        east=east,
        west=west,
        network_type="drive",
        simplify=True,
    )

    fig, ax = plt.subplots(figsize=(14, 14), facecolor="white")

    ox.plot_graph(
        graph,
        ax=ax,
        show=False,
        close=False,
        bgcolor="white",
        edge_color="#b8b8b8",
        edge_linewidth=0.8,
        node_size=0,
    )

    ax.set_facecolor("white")

    for _, row in df.iterrows():
        ax.scatter(
            row["lot"],
            row["lan"],
            color=color_map.get(str(row.get("point_type", "")).strip()),
            s=22,
            edgecolors="white",
            linewidths=0.4,
            zorder=5,
        )

    handles = [
        plt.Line2D(
            [0], [0],
            marker="o",
            color="w",
            label=point_type,
            markerfacecolor=color,
            markeredgecolor="black",
            markersize=8,
        )
        for point_type, color in color_map.items()
    ]

    ax.legend(handles=handles, title="Point Types", loc="upper right", frameon=True)

    ax.set_xlim(west, east)
    ax.set_ylim(south, north)

    output_png = str(Path(csv_path).with_name(f"{Path(csv_path).stem}_colored_osmnx_map.png"))
    plt.savefig(output_png, dpi=300, bbox_inches="tight", facecolor="white")
    plt.close(fig)

    return output_png
```

```python
kyiv_image = create_colored_map_osmnx('synthetic_data_kyiv_varash/kyiv/kyiv_day_03.csv')
```
